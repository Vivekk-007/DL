```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout,BatchNormalization
from tensorflow.keras.optimizers import Adam
```

```python
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")

print(train.shape)
```

```python
X = train.drop("label", axis=1).values
y = train["label"].values
```

```python
X = X / 255.0
test_X = test.values / 255.0

X = X.reshape(-1, 28, 28, 1)
test_X = test_X.reshape(-1, 28, 28, 1)
```

```python
X_train, X_val, y_train, y_val = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
model = Sequential([
    Conv2D(32, (3,3),strides=(2,2),padding="same",activation='relu', input_shape=(28,28,1)),
    MaxPooling2D(2,2),

    Conv2D(64, (3,3),strides=(2,2),padding='same', activation='relu'),
    BatchNormalization(),
    MaxPooling2D(2,2),
    Dropout(0.25),

    Conv2D(128, (3,3),strides=(2,2),padding='same', activation='relu'),
    BatchNormalization(),
    Dropout(0.20),

    Conv2D(128, (3,3),strides=(2,2),padding='same', activation='relu'),
    BatchNormalization(),
    Dropout(0.20),

    Flatten(),
    Dense(512, activation='relu'),
    BatchNormalization(),
    Dropout(0.30),

    Dense(256, activation='relu'),
    BatchNormalization(),
    Dropout(0.30),

    Dense(128, activation='relu'),
    BatchNormalization(),
    Dropout(0.30),

    Dense(64, activation='relu'),
    BatchNormalization(),
    Dropout(0.30),

    Dense(10, activation='softmax')
])

model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()
```

```python
history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=30,
    batch_size=128
)
```

```python
predictions = model.predict(test_X)
pred_labels = np.argmax(predictions, axis=1)

print(len(pred_labels))
```

```python
submission = pd.DataFrame({
    "ImageId": np.arange(1, 28001),
    "Label": pred_labels
})

submission.to_csv("submission.csv", index=False)

```
